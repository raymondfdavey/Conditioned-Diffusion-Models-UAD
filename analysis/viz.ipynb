```python
%matplotlib inline
import nibabel as nib
import matplotlib.pyplot as plt
from mri_viewer import get_valid_files
import os

# Get list of valid files
data_dir = "/home/rd81/projects/data-complete/Data/Test/Brats21"
t1_files = get_valid_files(os.path.join(data_dir, "t1"))

# Load and display first file
file_path = t1_files[0]
img = nib.load(file_path).get_fdata()
print(f"Loaded {os.path.basename(file_path)}")
print(f"Image shape: {img.shape}")

# Display middle slice
middle_slice = img.shape[2] // 2
plt.figure(figsize=(10, 5))
plt.imshow(img[:, :, middle_slice], cmap='gray')
plt.title(f'T1 Image - {os.path.basename(file_path)} - Layer {middle_slice}')
plt.axis('off')
plt.show()
```
